```python
#Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Load built-in dataset
df = sns.load_dataset("tips")  # Total bill, tip, day, gender, time etc.

# 1. Line Plot
sns.lineplot(x="total_bill", y="tip", data=df)
plt.title("Line Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.grid(True)
plt.show()

# 2. Bar Plot (Average tip by day)
sns.barplot(x="day", y="tip", data=df, palette="viridis")
plt.title("Bar Plot: Avg Tip per Day")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.grid(True)
plt.show()

# 3. Count Plot (Number of people per day)
sns.countplot(x="day", data=df, palette="Set2")
plt.title("Count Plot: Number of Customers per Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.grid(True)
plt.show()

# 4. Box Plot (Distribution of tips by gender)
sns.boxplot(x="sex", y="tip", data=df, palette="coolwarm")
plt.title("Box Plot: Tip Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Tip")
plt.grid(True)
plt.show()

# 5. Violin Plot (Detailed distribution of tips by gender)
sns.violinplot(x="sex", y="tip", data=df, palette="Pastel1")
plt.title("Violin Plot: Tip Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Tip")
plt.grid(True)
plt.show()

# 6. Histogram (Total bill)
sns.histplot(df["total_bill"], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram: Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# 7. KDE Plot (Tip density curve)
sns.kdeplot(df["tip"], shade=True, color="purple")
plt.title("KDE Plot: Tip Density")
plt.xlabel("Tip")
plt.grid(True)
plt.show()

# 8. Scatter Plot (Bill vs Tip)
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex", style="smoker")
plt.title("Scatter Plot: Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.grid(True)
plt.show()

# 9. Heatmap (Correlation between numeric columns)
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap: Feature Correlation")

plt.show()

# 10. Pair Plot (Auto matrix of scatter + hist)
sns.pairplot(df, hue="sex")
plt.suptitle("Pair Plot of All Numeric Features", y=1.02)
plt.show()
```
